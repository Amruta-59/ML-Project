```python
df = df.drop(["Patient_ID", "OCR_Extracted_Text"], axis=1)
```

```python
df
```

```python
import sklearn as sk
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
le = LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
df["Sleep_Disorder_Type"]=le.fit_transform(df["Sleep_Disorder_Type"])
```

```python
df
```

```python
x=df.drop(columns="Diagnosis_Confirmed")
```

```python
y=df["Diagnosis_Confirmed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=50)
```

```python
scaler=StandardScaler()
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test)
```

### Logistic Regression

```python
model=LogisticRegression()
```

```python
model.fit(x_train_scale,y_train)
```

```python
y_pred = model.predict(x_test_scale)
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
precision=precision_score(y_test,y_pred)
```

```python
precision
```

```python
recall=recall_score(y_test,y_pred)
```

```python
recall
```

```python
f1=f1_score(y_test,y_pred)
```

```python
f1
```

```python
report=classification_report(y_test,y_pred)
```

```python
print(report)
```
